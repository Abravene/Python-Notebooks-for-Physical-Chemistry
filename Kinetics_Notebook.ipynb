# 1 - First-Order Reactions

In the first-order reaction the rate is directly proportional to the concentration of a single species. Consider the reaction,

$$A \xrightarrow{k} B$$


The rate of the reaction is given by

$$rate = \frac{-d[A]}{dt}= \frac{d[B]}{dt} = k[A]$$


Let's find the integrated rate law with respect to $[A]$,

$$\frac{d[A]}{dt} = -k[A]$$

By separation of variables,

$$\frac{d[A]}{[A]} = −k dt$$


Integrating we have,

$$\int_{[A]_0}^{[A]_t} \frac{1}{[A]}d[A] = -k\int_{0}^{t}dt$$


The integrated equation can be written in the following way for convenience,

$$ln [A]_t = ln [A]_0 − kt$$

and

$$[A]_t = [A]_0e^{−kt}$$


For the calculation of a first-order rate constant, knowing the concentration of a reactant or product is not required. Also, it is not necessary to use a very high purity reagent in the case that the impurity does not affect the kinetics.

#2 - Half-life
The half-life, $t_{1/2}$, of a reaction is the time required for the concentration of a given reactant to reach a value that is the arithmetic mean of its initial and final, or equilibrium, values. Or more simply, it is the time when a quantity is reduced to half of its initial value. So for a reactant that is entirely consumed, it is the time required for the reactant concentration to fall to one-half of its initial value.

The half-life of a reaction has an exact quantitative value for a few specific situations:

1. for a first-order reaction where the half-life of the reactant can be called the half-life of the reaction; 

2. for a reaction involving more than one reactant, with their concentrations in stoichiometric ratios. In this case, the half-life of each reactant is the same and may be called the half-life of the reaction. 

Note that if the concentrations of reactants are not in
stoichiometric ratios, the half-lives for the different reactants are not the same and use of the term half-life of the reaction is not strictly appropriate.


Let's consider the half-life for the zero order reaction, 


$$rate = \frac{-d[A]}{dt}= \frac{d[B]}{dt} = k$$

which is given by 

$$t_{1/2}=\frac{[A]_0}{2k}$$

Note that the half-life of  a zero order reaction depends only on the initial concentration of reactant and the rate constant.

# 3 - Lifetime ($\tau$) and e-folding

Now that we have considered first order reactions and the concept of a half-life, let's connect the two ideas. To do so, we will consider the chemical lifetime ($\tau$). The chemical lifetime of a molecular entity reacting with first-order kinetics is defined as the time required for its concentration to decrease to 1/e (≈0.37) of its initial value. 

The term lifetime is commonly used in atomospheric chemistry. For example, the greenhouse gas tetrafluoromethane $CF_4$ has an atmospheric lifetime of approximately 50,000 years, while methane $CH_{4}$ has one of 12 years. The term is also used in NMR analysis where you may consider the lifetimes of the magnetic states between which the NMR transitions occur. 

Statistically, lifetime represents the life “expectancy” of the entity. It is equal to the reciprocal of the sum of the first-order rate constants of all processes causing the decay of the molecular entity.

For a first order reaction,

$$\frac{[A]_t}{[A]_0} = e^{−kt}$$

The units for k are inverse time, 1/s. If we were to take the reciprocal of this, we would be left with units of time. 

For this reason, we often state that the lifetime of a species that undergoes first order decay is equal to the reciprocal of $k$. Consider, now, what would happen if we were to set the time, $t$, to the reciprocal of the rate constant, $k$ such that,


$$t=1/k$$

then

$$\frac{[A]_t}{[A]_0} = e^{−k(1/k)}=e^{(-1)}\approx 0.37$$


We have lost (1 − 0.37 = 0.63) 63% of A, with only 37% left. With this, we now know that if 1 lifetime has occured, we have gone through 1 e-folding.

Likewise for an e-folding of two

$$\frac{[A]_t}{[A]_0} =e^{−k(2/k)}=e^{(-2)}\approx 0.135$$
