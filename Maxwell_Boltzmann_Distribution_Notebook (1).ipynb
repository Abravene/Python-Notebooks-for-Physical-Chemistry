#Intro: The Maxwell-Boltzmann Distribution

In the figure below, the x-axis displays the speeds of the molecules. The y-axis displays $f(v)$ which represents the fraction of molecules. A higher curve at a given speed means more molecules travel at that speed.

In the figure below, it can be seen that the overwhelming majority of argon molecules have speeds between 0 and 2000 m/s for temperatures from 25 to 1200 K. 

Note that the speeds are not limited to the range shown on this graph.

```
#Figure 1: Static Image

import numpy as np
import matplotlib.pyplot as plt
 
# Variables
m = 39.948*1.673534e-27 #Ar kg mol * dalton weighs: 1.008u or 1.673534e-27 kilograms
#Argon MW = 39.948 u
T1 = 25 #temperature 1
T2 = 100 #temperature 2
T3 = 300 #temperature 3
T4 = 1200 #temperature 4
kB = 1.3806503e-23 # m2 kg s-2 K-1 #the Boltzmann constant


#Argon

def fm1(m, T1, v):
    return 4.*np.pi*v*v*((m/(2.*np.pi*kB*T1))**1.5) * np.exp((-m*v*v)/(2.*kB*T1))
 
def fm2(m, T2, v):
    return (m/(2.*np.pi*kB*T2))**1.5 * 4.*np.pi*v*v * np.exp((-m*v*v)/(2.*kB*T2))
    
def fm3(m, T3, v):
    return (m/(2.*np.pi*kB*T3))**1.5 * 4.*np.pi*v*v * np.exp((-m*v*v)/(2.*kB*T3))

def fm4(m, T4, v):
    return (m/(2.*np.pi*kB*T4))**1.5 * 4.*np.pi*v*v * np.exp((-m*v*v)/(2.*kB*T4))


v = np.arange(0.0, 2000, 0.01) #set graph limits from 0 to 2000 K
 
 
#Plot the figure

plt.figure(figsize=(7,7))
plt.text( 600, 0.005,r'$f(v) = \sqrt{(\frac{m}{2 \pi k_{B} T})^{3}} 4\pi v^{2} \exp{\frac{-mv^{2}}{2k_{B}T}}$', fontsize=18)
plt.plot(v, fm1(m, T1, v), 'crimson', label=str(T1)+' K') 
plt.plot(v, fm2(m, T2, v), color='darkorchid', label=str(T2)+' K')
plt.plot(v, fm3(m, T3, v), 'b', alpha=0.7, label=str(T3)+' K') 
plt.plot(v, fm4(m, T4, v), 'b', alpha=0.5, label=str(T4)+' K') 

plt.fill_between(v, fm1(m, T1, v,), alpha=0.15, color='crimson')
plt.fill_between(v, fm2(m, T2, v,), alpha=0.1, color='magenta')
plt.fill_between(v, fm3(m, T3, v,), alpha=0.1, color='b')
plt.fill_between(v, fm4(m, T4, v,), alpha=0.1, color='cyan')

plt.xlabel('v (m/s)', fontsize=14)
plt.ylabel('f(v)', fontsize=16)

plt.title('Maxwell-Boltzmann speed distribution for Ar')

plt.legend()
plt.show()
```

#2 - Most Probable, Mean, and Root Mean Square Speed

The speed that corresponds to the peak of the curve is called the 
most probable speed. More molecules travel at (or close
to) this speed than any other.

The most probable speed is given by the formula

$v_{mp}=\sqrt{\frac{2RT}{M}}.$

where T = the kelvin temperature\
R = 8.314 J/mol K\
M = the molar mass in kg/mol.

The mean speed is given by,

$v_{mean}=\sqrt{\frac{8RT}{\pi M}}.$

Finallly, the root-mean-square speed is the speed that corresponds to
the average kinetic energy of the molecules, 

$v_{rms}=\sqrt{\frac{3RT}{M}}.$


In fact the root-meansquare speed is 22.5% higher than the
most probable speed and 8.5% higher than the average speed.

```
#Most Probable, Mean, and Root Mean Square Speed

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets, interact, interactive
from IPython.display import clear_output, display, HTML

kB = 1.3806503e-23 # m2 kg s-2 K-1
R=8.314
M=0.039948 #Argon

def fm1(T):

    fm1=[]
    v = np.arange(1500)
    for ii in np.arange(len(v)): fm1.append((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*v[ii]*v[ii] * np.exp((-M*v[ii]*v[ii])/(2.*R*T)))
    

    plt.figure(figsize=(7,7))
    plt.xlabel('v (m/s)', fontsize=14)
    plt.ylabel('f(v)', fontsize=16)

    plt.plot(v,fm1, alpha=0.9, color='black')

    plt.text( 600, 0.005, '$v_{p}<v_{mean}<v_{rms}$', fontsize=18, )

    vp=((2.*R*T)/M)**0.5
    vrms=((3.*R*T)/M)**0.5
    vmean=((8.*R*T)/(M*np.pi))**0.5

    fvp =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vp*vp * np.exp((-M*vp*vp)/(2.*R*T)))
    fvrms =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vrms*vrms * np.exp((-M*vrms*vrms)/(2.*R*T)))
    fmean =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vmean*vmean * np.exp((-M*vmean*vmean)/(2.*R*T)))


    plt.plot([vp,vp],[0,fvp], marker=',', color='r', label='$v_{mp}$: '+'{:5.2f}'.format(vp)+ ' m/s')
    plt.plot([vmean, vmean],[0,fmean],marker=',', label='$v_{mean}$: '+'{:5.2f}'.format(vmean)+ ' m/s')
    plt.plot([vrms, vrms],[0,fvrms],marker=',', label='$v_{rms}$: '+'{:5.2f}'.format(vrms)+ ' m/s')
    

    #plt.plot([0,1],[0,1],'--', color='gray', label='half-life')

    #plt.fill_between(v, fm1, alpha=.2, color='springgreen')

    plt.title('Maxwell-Boltzmann speed distribution for Ar'+'       T = ' +str(T)+' K')
    plt.ylim(0, 0.009)

    plt.legend()
    

interactive_plot = interactive(fm1,T=(25,1600,100), label='name')
interactive_plot
```

```
#The Maxwell-Boltzmann distribution is normalized

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets, interact, interactive
from IPython.display import clear_output, display, HTML
from scipy.integrate import trapz, simps


kB = 1.3806503e-23 # m2 kg s-2 K-1
R=8.314
M=0.016 #CH4 in kg mol-1


def fm1(T):

    fm1, fm800=[],[]
    
    v = np.arange(2000)
    v800 = np.arange(800)

    for ii in np.arange(len(v)): fm1.append((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*v[ii]*v[ii] * np.exp((-M*v[ii]*v[ii])/(2.*R*T)))

    for ii in np.arange(len(v800)): fm800.append((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*v[ii]*v[ii] * np.exp((-M*v[ii]*v[ii])/(2.*R*T)))

    plt.figure(figsize=(7,7))
    plt.xlabel('v (m/s)', fontsize=14)
    plt.ylabel('f(v)', fontsize=16)

    plt.plot(v,fm1, alpha=0.9, color='black')

    plt.text(600, 0.005, 'Area under curve = 1', fontsize=18, )

    #This is where you may change the integration method from trapz to simps
    plt.text(600, 0.0035,'Percent of molecules with velocities' +'\n'+ 'between 0 to 800 m/s: '+str('{:5.2f}'.format(trapz(fm800, v800)*100)+'%'), fontsize=12)

    vp=((2.*R*T)/M)**0.5
    vrms=((3.*R*T)/M)**0.5
    vmean=((8.*R*T)/(M*np.pi))**0.5

    fvp =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vp*vp * np.exp((-M*vp*vp)/(2.*R*T)))
    fvrms =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vrms*vrms * np.exp((-M*vrms*vrms)/(2.*R*T)))
    fmean =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vmean*vmean * np.exp((-M*vmean*vmean)/(2.*R*T)))


    plt.plot([vp,vp],[0,fvp], marker=',', color='r', label='$v_{mp}$: '+'{:5.2f}'.format(vp)+ ' m/s')
    plt.plot([vmean, vmean],[0,fmean],marker=',', label='$v_{mean}$: '+'{:5.2f}'.format(vmean)+ ' m/s')
    plt.plot([vrms, vrms],[0,fvrms],marker=',', label='$v_{rms}$: '+'{:5.2f}'.format(vrms)+ ' m/s')

    plt.xlim(0, 2000)
    plt.ylim(0, 0.009)

    plt.fill_between(v800, fm800, alpha=0.2, color='gray', label='% molecules')
    
    plt.title('Maxwell-Boltzmann speed distribution'+'   T = ' +str(T)+' K')
    plt.legend()
    
interactive_plot = interactive(fm1,T=(25,2000,100), label='name')
interactive_plot
```

#4 - Kinetic Energy

It is also possible to plot the kinetic energy distribution for gas particles. As before, the total area under this curve is 1. Therefore, integrating the area under the curve gives the fraction of molecules within a certain kinetic energy range. 

The most probable kinetic energy corresponds to the peak and is given by the formula,

$KE_{mp}=\frac{1}{2}RT$.

The largest fraction of molecules have this kinetic energy; thus, it is the most probable kinetic energy.

The mean kinetic energy is given by the formula,

$KE_{mean}=\frac{3}{2}RT$.

Since the distribution is not symmetrical, the most probable speed is not the same as the mean speed. The average kinetic energy is much larger than the most probable kinetic energy. 

$KE_{mean} > KE_{mp}$

Note that for the speed distributions, the most probable speed was fairly close to the average speed.

```
#Kinetic Energy

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets, interact, interactive
from IPython.display import clear_output, display, HTML
from scipy.integrate import trapz, simps


kB = 1.3806503e-23 # m2 kg s-2 K-1
R=8.314


def fm1(T):

    fm1=[]
    KE = np.arange(200000)
    for ii in np.arange(len(KE)): fm1.append(2*np.pi*((1/(np.pi*R*T))**(3/2))*(KE[ii]**(0.5))*np.exp(-KE[ii]/(R*T)))

    fm40k=[]
    KE40k = np.arange(20000, 40000)
    for ii in np.arange(len(KE40k)): fm40k.append(2*np.pi*((1/(np.pi*R*T))**(3./2.))*(KE40k[ii]**(0.5))*np.exp(-KE40k[ii]/(R*T)))

    plt.figure(figsize=(7,7))
    plt.xlabel('Kinetic Energy (J/mol)', fontsize=14)
    plt.ylabel('Fraction per J/mol', fontsize=16)
    
    plt.plot(KE, fm1, alpha=0.8, color='black')
    plt.ylim(0, 0.0002)
    plt.xlim(0, 40000)

    KE_mp = (1/2)*R*T
    fKE_mp =(2*np.pi*((1/(np.pi*R*T))**(3/2))*(KE_mp**(0.5))*np.exp(-KE_mp/(R*T)))
    plt.plot([KE_mp, KE_mp],[0,fKE_mp],marker=',', label='$KE_{mp}$: '+'{:5.2f}'.format(KE_mp)+ ' J/mol')

    KE_mean = (3/2)*R*T
    fKE_mean =(2*np.pi*((1/(np.pi*R*T))**(3/2))*(KE_mean**(0.5))*np.exp(-KE_mean/(R*T)))
    plt.plot([KE_mean, KE_mean],[0,fKE_mean],marker=',', label='$KE_{mean}$: '+'{:5.2f}'.format(KE_mean)+ ' J/mol')
    
    plt.text(13000, 0.00012,'Percent of molecules with kinetic energy' +'\n'+ 'above $E_{a}$ of 20,000 J/mol: '+str('{:5.2f}'.format(trapz(fm40k, KE40k)*100)+'%'), fontsize=12)
    plt.fill_between(KE40k, fm40k, alpha=0.2, color='gray')
    plt.title('Maxwell-Boltzmann kinetic energy distribution'+'   T = ' +str(T)+' K')

    plt.legend()


interactive_plot = interactive(fm1,T=(298.15,1000,100), label='name')
interactive_plot

```

#5 -  The Speed of Sound 

The speed of sound is dependant on both the temperature and the medium through which a sound wave is propagating. Let's investigate both.


The root mean square speed is directly related to the speed of sound c in the gas, by

$c={\sqrt {\frac {\gamma }{3}}}$

$ v_{\mathrm {rms} }={\sqrt {\frac {f+2}{3f}}}\ $

$v_{\mathrm {rms} }={\sqrt {\frac {f+2}{2f}}}\ v_{p}$

where ${\textstyle \gamma =1+{\frac {2}{f}}}{\textstyle \gamma =1+{\frac {2}{f}}}$ is the adiabatic index (defined as the ratio of the heat capacity at constant pressure to heat capacity at constant volume), $f$ is the number of degrees of freedom of the individual gas molecule. 


Earth's atmosphere is composed of about 78 percent nitrogen, which we can use to approximate air. For diatomic nitrogen at 300 K, ${\displaystyle f=5}{\displaystyle f=5}$ so


${\displaystyle c={\sqrt {\frac {7}{15}}}v_{\mathrm {rms} }\approx 68\%\ v_{\mathrm {rms} }\approx 84\%\ v_{p}.}$

```
#Speed of Sound

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets, interact, interactive
from IPython.display import clear_output, display, HTML
from scipy.integrate import trapz, simps


kB = 1.3806503e-23 # m2 kg s-2 K-1
R=8.314
M = 0.028 #molar mass for N2 in kg/mol
M_air = 0.02897 #the average molar mass of dry air is 28.97 g/mol


def fm1(T):

    fm1, fm_air=[],[]
    
    """ Calculates the adiabatic index of air at temperatures ranging 
    233.15 - 673.15 K at standard atmospheric pressure """

    K_air=0.0000000003*(T)**3 - (0.0000005)*(T)**2 + 0.0002*T + 1.3768

    """ Generates MB curves """
    v = np.arange(2000)
    for ii in np.arange(len(v)): fm1.append((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*v[ii]*v[ii] * np.exp((-M*v[ii]*v[ii])/(2.*R*T)))
    for ii in np.arange(len(v)): fm_air.append((M_air/(2.*np.pi*R*T))**1.5 * 4.*np.pi*v[ii]*v[ii] * np.exp((-M_air*v[ii]*v[ii])/(2.*R*T)))

    plt.figure(figsize=(7,7))
    plt.xlabel('v (m/s)', fontsize=14)
    plt.ylabel('f(v)', fontsize=16)

    plt.plot(v,fm_air, alpha=0.9, color='black',  label='air')
    plt.plot(v,fm1, alpha=0.9, color='gray', linestyle = 'dashed', label='$N_{2}$: ')

    vp=((2.*R*T)/M)**0.5 
    vrms=(((3.*R*T)/M)**0.5)
    vrms_air=(((3.*R*T)/M_air)**0.5) #uses MW of dry air
    vmean=((8.*R*T)/(M*np.pi))**0.5

    c=((K_air/3)**0.5)*vrms_air #this is for air

    vp_C=(((2.*R*T)/M)**0.5)*(0.84) #this will approximate c in m/s
    vrms_C=(((3.*R*T)/M)**0.5)*(0.68) #this will approximate c in m/s
    avg_C=(vp_C+vrms_C)/2 #finds the mean of the two above lines

    fvp =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vp*vp * np.exp((-M*vp*vp)/(2.*R*T)))
    fvrms =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vrms*vrms * np.exp((-M*vrms*vrms)/(2.*R*T)))

    fvp_C =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vp_C*vp_C* np.exp((-M*vp_C*vp_C)/(2.*R*T)))
    fvrms_C =((M/(2.*np.pi*R*T))**1.5 * 4.*np.pi*vrms_C*vrms_C * np.exp((-M*vrms_C*vrms_C)/(2.*R*T)))

    plt.text(800, 0.0015, 'Speed of Sound for ' + '$N_{2}$,' +'\n'+ 'c '+str('{:5.2f}'.format(avg_C)+' m/s'), fontsize=18, )
    plt.text(900, 0.001, 'Using  of air,' +'\n'+ 'c '+str('{:5.2f}'.format(c)+' m/s'), fontsize=18, )

    plt.plot([vp_C,vp_C],[0,fvp_C], marker=',', color='r', label='84$\%$ $v_{mp}$ for $N_{2}$: '+'{:5.2f}'.format(vp_C)+ ' m/s')
    plt.plot([vrms_C, vrms_C],[0,fvrms_C],marker=',', label='68$\%$ $v_{rms}$ for $N_{2}$: '+'{:5.2f}'.format(vrms_C)+ ' m/s')

    plt.xlim(0, 2000)
    plt.ylim(0, 0.0025)
    
    plt.title('Maxwell-Boltzmann speed distribution'+'   T = ' +str(T)+' K')
    plt.legend()
  

interactive_plot = interactive(fm1,T=(240,350,10), label='name')
interactive_plot

```
